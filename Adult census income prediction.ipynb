## Feature Engineering

```python
for dataset in [df]:
    dataset.loc[dataset['country'] != ' United-States', 'country'] = 0
    dataset.loc[dataset['country'] == ' United-States', 'country'] = 1
    dataset.loc[dataset['race'] != ' White', 'race'] = 0
    dataset.loc[dataset['race'] == ' White', 'race'] = 1
    dataset.loc[dataset['workclass'] != ' Private', 'workclass'] = 0
    dataset.loc[dataset['workclass'] == ' Private', 'workclass'] = 1
    dataset.loc[dataset['hours-per-week'] <= 40, 'hours-per-week'] = 0
    dataset.loc[dataset['hours-per-week'] > 40, 'hours-per-week'] = 1
for col in df[df.columns]:
    if df[col].dtypes == 'object':
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
df = df.astype(int)
df=df.drop(["education"],axis=1)
df.head()
```

### train_test_split

```python
x = df.drop('salary',axis=1)
y = df['salary']
```

```python
df['salary'].value_counts()
```

##### As we see here, our data is imbalanced.Imbalanced data is a common problem in machine learning, which brings challenges to feature correlation, class separation and evaluation, and results in poor model performance.Classes that make up a large proportion of the data set are called majority classes. Those that make up a smaller proportion are minority classes. so by using randomoversampler method we make our data balanced.

```python
random_sampler = RandomOverSampler(random_state=30,sampling_strategy=0.95)
random_sampler.fit(x,y)
```

```python
x_new,y_new = random_sampler.fit_resample(x, y)
```

```python
y_new.value_counts()
```

```python
x_new_train, x_new_test, y_new_train,y_new_test = train_test_split(x_new,y_new,test_size=0.3,random_state=15,stratify=y_new)
x_new_train.shape, x_new_test.shape, y_new_train.shape, y_new_test.shape
```

## Model training

# Random_forest

```python
rf_model = RandomForestClassifier(random_state=15)
rf_model.fit(x_new_train,y_new_train)
```

```python
y_pred = rf_model.predict(x_new_test)
```

## Model evaluation

```python
# Training data Evalution 
def model_eval(model,x,y):
    y_pred = model.predict(x)

    cm = confusion_matrix(y, y_pred)
    print(f"Confusion Matrix = \n{cm}")
    print("*"*50)
    
    ac = accuracy_score(y, y_pred)
    print(f"Accuracy Score = {ac}")
    print("*"*50)

    cr = classification_report(y, y_pred)
    print(f"Classification Report = \n{cr}")
    
    pr = precision_score(y,y_pred)
    print(f'precision_score = {pr}')
    
    rc = recall_score(y,y_pred)
    print(f'recall_score = {rc}')
    
    f1 = f1_score(y,y_pred)
    print(f'f1_score = {f1}')
    
    return "Model Eval Success"
```

```python
model_eval(rf_model,x_new_train,y_new_train)
```

```python
model_eval(rf_model,x_new_test,y_new_test)
```

```python
cm = confusion_matrix(y_new_test, y_pred)
plt.figure(figsize=(6,4))
sns.heatmap(cm,annot=True,fmt='d',cmap="Blues")
plt.xlabel("Predicted Values")
plt.ylabel("Actual values")
plt.show()
```

#### In random forest algorithm, it consist low bias and low variance, so it is generalised model. it doesn't need of hyperparamter tuning

##### To finding the different values of k and p

```python
distanc_param=1 
train_acc_list = []
test_acc_list = []
k_range = np.arange(3,20)
for k in k_range:
    model = KNeighborsClassifier(n_neighbors=k,p=distanc_param)
    model.fit(x_new_train,y_new_train)
    # Training Acc
    train_acc = model.score(x_new_train,y_new_train)
    train_acc_list.append(train_acc)
    # Testing Acc
    test_acc_list.append(model.score(x_new_test,y_new_test))
print(train_acc_list)
print(test_acc_list)
```

```python
plt.plot(k_range,train_acc_list)
plt.plot(k_range,test_acc_list)
plt.xlabel("K Values ")
plt.ylabel("Accuracy")
plt.title("Train and Test Accuracy with Manhattan Distance")
plt.savefig("Accuracy_MD.png")
```

# Adaboost with different max_depth Decision tree

```python
train_acc_list = []
test_acc_list = []
for i in range(1,10):
    base_model = DecisionTreeClassifier()
    base_model = DecisionTreeClassifier(max_depth=i,random_state=7)
    model = AdaBoostClassifier(base_estimator=base_model,
                               n_estimators=50,
                              learning_rate=1.0,
                              random_state=10)

    model.fit(x_new_train,y_new_train)

    train_acc = model.score(x_new_train,y_new_train)
    train_acc_list.append(train_acc)

    test_acc = model.score(x_new_test,y_new_test)   
    test_acc_list.append(test_acc)
    
    print(f"Iteration >> {i}, Training Acc = {train_acc}, Testing Acc = {test_acc}")
```
